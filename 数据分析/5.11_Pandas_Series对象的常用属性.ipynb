```python
import pandas as pd
```

```python
# 1. 创建series对象
# s1 = pd.Series(data=[1,2,3,4,5,6], index=['a','b','c','d','e','f'])
# 加入列表推导式
s1 = pd.Series(data=[i for i in range(6)], index=[i for i in 'abcdef'])
print(s1)
```

```python
# 2. 获取series对象的索引列的值
print(s1.index)
```

```python
# 3. 获取series对象的列值的值
print(s1.values)
```

```python
# 4. series根据 索引 获得 元素
print(s1['d'])
```
