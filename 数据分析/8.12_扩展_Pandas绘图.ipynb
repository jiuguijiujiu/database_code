```python
# 导包
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 1. 加载数据集, 并设置第一列为索引列
df = pd.read_csv('./数据集/winemag-data_first150k.csv', index_col = 0)
df.head()
```

```python
# 2. 查看数据集
df.info()
df.describe()
df.columns
```

```python
# 需求1：绘制图像，显示 产葡萄酒最多的10的产地信息
# 3. 分组统计
# 思路1：根据产地（province）分组，然后根据评分points（也可以是其他不含确实值的列）进行count统计
df.groupby('province', as_index = False).agg({'points':'count'}).sort_values(by='points', ascending = False).head(10)

# 思路2：value_counts()，计算值的个数，且会降序排序，等于groupby() + count() + sort_values()
df.province.value_counts().head(10)
```

```python
# 4. 针对于上述结果绘图
df.province.value_counts().head(10).plot.line()      # 折线图
# df.province.value_counts().head(10).plot(figsize = (10, 5), color = 'r')             # 折线图语法糖
df.province.value_counts().head(10).plot.bar(figsize = (10, 5), color = ['r', 'g', 'b', 'y', 'c', 'm', 'w', 'pink', 'gray'], fontsize = 15)         # 柱状图
```

```python
# 5. 最终写法
# step1:定义变量,记录参数,字典形式
text_kwargs = dict(figsize = (10, 5), color = ['r', 'g', 'b', 'y', 'c', 'm', 'w', 'pink', 'gray'], fontsize = 15)
print(text_kwargs)

# step2:绘图传参
df.province.value_counts().head(10).plot.bar(**text_kwargs)
```

```python
# 需求2:展示 葡萄酒最多的10个产地 占比
# 1. 计算葡萄酒最多的十个产地
df['province'].value_counts().head(10)

# 2. 计算每个产地占比 计算: 该产地数/全表总产地数
# len(df)       # 150930
df['province'].value_counts().head(10) / len(df)

# 3. 画图
(df['province'].value_counts().head(10) / len(df)).plot.bar(**text_kwargs)
```

```python
# 需求3: 展示 每个评分的葡萄酒种类数(个数)
# df.points.value_counts().sort_index().plot.line(**text_kwargs)      # 折线图
# df.points.value_counts().sort_index().plot.area(**text_kwargs)      # 面积图
df.points.value_counts().sort_index().plot.bar(**text_kwargs)         # 柱状图
```
