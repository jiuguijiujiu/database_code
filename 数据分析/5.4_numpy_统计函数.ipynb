# 1. numpy统计函数演示

```python
# 1. 定义numpy对象ndaray数组
arr1 = np.arange(12).reshape(3,4)
print(arr1)
```

```python
print(np.cumsum(arr1))
```

```python
print(np.sum(arr1))
print(np.sum(arr1, axis = 0))
print(np.sum(arr1, axis = 1))
print(np.mean(arr1))
print(np.argmax(arr1))
print(np.argmin(arr1))
print(np.std(arr1))
```
