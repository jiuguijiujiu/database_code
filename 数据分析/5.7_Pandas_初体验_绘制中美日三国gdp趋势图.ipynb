```python
import pandas as pd
```

```python
# 1. 读取数据
# 绝对路径
# df = pd.read_csv(r'D:\code\黑马数据分析\数据分析\数据集\1960-2019全球GDP数据.csv', encoding='gbk')

# 相对路径
df = pd.read_csv(r'./数据集/1960-2019全球GDP数据.csv', encoding='gbk')
```

```python
# 2. 加载中国数据
chain_df = df[df.country == '中国']
chain_df
```

```python
# 3. 设置year字段为索引列
chain_df.set_index('year', inplace=True)        # inplace=True表示修改源数据
chain_df
```

```python
# 4. 绘制折线图
# print(chain_df.plot())
chain_df.GDP.plot()
```

```python
# 5. 参考上面思路绘制中美日三国gdp
usa_df = df[df.country == '美国'].set_index('year')
jp_df = df[df.country == '日本'].set_index('year')
usa_df
jp_df
```

```python
# 6. 绘制三国折线图
chain_df.GDP.plot()
usa_df.GDP.plot()
jp_df.GDP.plot()
```
