```python
import pandas as pd
```

```python
# 1. 准备数据
df1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                    'key2': ['K0', 'K1', 'K0', 'K1'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})

df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                    'key2': ['K0', 'K0', 'K0', 'K0'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

# 2. 查看数据
df1
df2
```

```python
# 3. 演示merge()默认合并方式
pd.merge(df1, df2)
pd.merge(df1, df2, how = 'inner', on = ['key1', 'key2'])   # 效果同上
```

```python
# 4. 内连接，指定合并字段
pd.merge(df1, df2, how = 'inner', on = 'key2')
```

```python
# 5. 外连接，指定合并字段
pd.merge(df1, df2, how = 'outer', on = ['key1', 'key2'])

# 左外连接
pd.merge(df1, df2, how = 'left', on = ['key1', 'key2'])
```

```python
# 6. merge()函数其他写法
df1.merge(df2, how = 'inner', on = 'key1')  # 效果同上

# 回顾concat()
# df1.concat(df2)       # 错误写法
pd.concat([df1, df2, df1])
```
