```python
import matplotlib.pyplot as plt
import random

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 需求：在一个画布中 绘制 两个城市的温度图

# 1. x，y轴数据准备
x = range(60)
y_beijing = [random.uniform(15, 20) for i in x]
y_shanghai = [random.uniform(5, 10) for i in x]

# 2. 准备画布
plt.figure(figsize=(15, 5), dpi=80)

# 3. 画图
plt.plot(x, y_beijing, 'red', label='beijing')
plt.plot(x, y_shanghai, 'blue', linestyle = '--', label='shanghai')

# 4. 设置网格
plt.grid(linestyle = '--', alpha = 0.5)

# 5. 设置标题与x,y轴标签
plt.title('北京与上海的11点——12点温度', fontsize = 20, color = 'green')
plt.xlabel('时间', fontsize = 15)
plt.ylabel('温度', fontsize = 15)

# 6. 设置x，y轴刻度
# 6.1 刻度数据准备
x_ticks_label = [f'11点{i}分' for i in x]
y_ticks = range(40)
y_ticks_label = [f'{i}度' for i in y_ticks]
# 6.2 设置刻度
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5], y_ticks_label[::5])

# 7. 设置
plt.legend(loc = 0)

# 8. 保存
plt.savefig('./数据集/my_picture1.png')

# 4. 展示
plt.show()
```
