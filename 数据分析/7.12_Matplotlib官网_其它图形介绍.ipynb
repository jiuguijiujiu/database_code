```python
import matplotlib.pyplot as plt
import numpy as np
import random

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 1. 柱状图bar()——>适用于对比，统计

```python
# 0. 准备数据
x = ['A', 'B', 'C', 'D']
y = [10, 20, 30, 40]

# 1. 创建画布
plt.figure(figsize=(10, 5), dpi=100)

# 2. 绘制柱状图
plt.bar(x, y, color='c')

# 设置网格
plt.grid(linestyle = '--', alpha = 0.5)

# 设置标题，x与y轴标签
plt.title('柱状图', fontsize = 20)
plt.xlabel('分类名', fontsize = 10)
plt.ylabel('数量', fontsize = 10)

# 3. 展示结果
plt.show()
```

# 2. 直方图hist()——>谁用于 连续数据的 分布和统计

```python
# 0. 准备数据
data = np.random.randn(500)            # 生成500个随机数，符合正态分布

# 1. 创建画布
plt.figure(figsize=(10, 5), dpi=100)

# 2. 绘制直方图
plt.hist(data, bins=10, color='c')

# 设置网格
plt.grid(linestyle = '--', alpha = 0.5)

# 设置标题，x与y轴标签
plt.title('直方图', fontsize = 20)

# 3. 展示结果
plt.show()
```

# 3. 饼图pie()——>适用于 比例 的统计

```python
sizes = [25, 35, 25, 15]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

# 1. 创建画布，绘制饼图
plt.pie(sizes, labels=labels, autopct='%1.1f%%') # 绘制饼图，显示百分比


plt.title("Simple Pie Chart") # 设置图表标题


plt.show() # 显示图表
```

# 4. 散点图scatter()——>使用与 散点数据的 统计，或统计两列值之间的关系

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='red') # 绘制红色散点图


plt.title("Simple Scatter Plot") # 设置图表标题
plt.xlabel("X-axis") # 设置X轴标签
plt.ylabel("Y-axis") # 设置Y轴标签


plt.show() # 显示图表
```
