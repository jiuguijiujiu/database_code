```python
import numpy as np
```

# 1. 创建等比数列

```python
arr1 = np.logspace(0, 9, 10)  # base不写默认10
print(arr1)
print(arr1.dtype)
```

```python
arr2 = np.logspace(0, 9, 10, base = 2, dtype = np.int64)  # base不写默认10
print(arr2)
print(arr2.dtype)
```

# 2. 创建等差数列

```python
arr3 = np.linspace(1, 10, 4, endpoint = False, dtype = np.int64)
print(arr3)
print(arr3.dtype)
```
