```python
import pandas as pd
```

```python
# 1. 交叉表
# 创建一个模拟数据集
data1 = {
    '性别': ['男', '女', '男', '女', '男', '女', '女', '男'],
    '购买': ['是', '否', '是', '是', '否', '否', '是', '否'],
}

df1 = pd.DataFrame(data1)

# 创建交叉表
crosstab = pd.crosstab(df1['性别'], df1['购买'])
print(crosstab)
```

```python
# 2. 透视表完成上述需求（掌握）
# 创建一个模拟数据集
data2 = {
    '性别': ['男', '女', '男', '女', '男', '女'],
    '购买': ['是', '否', '是', '是', '否', '否'],
    '金额': [100, 150, 200, 130, 160, 120]
}

df2 = pd.DataFrame(data2)
df2.pivot_table(values = '金额', index = '性别', columns = '购买', aggfunc = 'mean')
```

```python
# 3. 透视表案例：优衣库数据集
# 3.1 读取数据
df3 = pd.read_csv('./数据集/uniqlo.csv')
df3

# 3.2 需求：根据城市，销售渠道分组，计算 销售金额 总和。
# 写法1：通用版本，分组聚合
df3.groupby(['city', 'channel']).agg({'revenue':'sum'})
df3.groupby(['city', 'channel'], as_index = False).agg({'revenue':'sum'})

# 写法2：pivot_table()透视表
df3.pivot_table(values = 'revenue', index = 'city', columns = 'channel', aggfunc = 'sum')
```
