```python
import pandas as pd
```

```python
# 1. 读取数据
df = pd.read_csv('./数据集/uniqlo.csv')
df
```

```python
# 2. 分组聚合，groupby([]).agg({})
# 场景1： 按照单列分组
df.groupby(['city'])
df.groupby('city')
```

```python
# 场景2： 按照多列分组
df.groupby(['city', 'channel'])
df.groupby(['city', 'channel']).revenue
```

```python
# 场景3：如何获取某个分组的数据
df.groupby(['city', 'channel']).get_group(('北京', '线下'))     # 一个分组信息

df.groupby(['city', 'channel']).get_group(('上海', '线上'))     # 另一个分组信息
```

```python
# 场景4：分组 + 聚合（聚合字段一个）
# 根据 城市， 销售渠道 分组， 求销售金额
# 写法1：通用版(掌握)
df.groupby(['city', 'channel']).agg({'revenue':'sum'})

# 写法2：变形版（了解）
df.groupby(['city', 'channel']).revenue.sum()
df.groupby(['city', 'channel'])['revenue'].sum()
df.groupby(['city', 'channel'])[['revenue']].sum()
```

```python
# 场景5：分组 + 聚合（聚合字段两个，聚合操作相同）
# 根据 城市， 销售渠道 分组， 求 总销售金额，总订单数列
# 写法1：通用版(掌握)
df.groupby(['city', 'channel']).agg({'revenue':'sum', 'order':'sum'})

# 写法2：变形版（了解）
df.groupby(['city', 'channel'])[['revenue', 'order']].sum()
```

```python
# 场景6：分组 + 聚合（聚合字段两个，聚合操作不同）
# 根据 城市， 销售渠道 分组， 分别求 销售金额的平均值 和 成本的总和
df.groupby(['city', 'channel']).agg({
    'revenue':'mean',
    'unit_cost':'sum'
})
```
