```python
import numpy as np
import pandas as pd
```

```python
# 1. 字典+列表创建df，使用默认自增索引
info = {
    'name':['一一', '二二', '三三', '四四'],
    'gender':['男', '女', '未知', '保密'],
    'age':[20, 30, 40, 50]
}

df1 = pd.DataFrame(data = info)
df1
```

```python
# 2. 列表+元组创建df，并自定义索引
info = [
    ('哈哈', '男', 18),
    ('呵呵', '女', 19),
    ('哇哇', '未知', 20),
    ('桀桀', '保密', 21)
]

df2 = pd.DataFrame(data = info, index= ['a', 'b', 'c', 'd'], columns = ['name', 'age', 'gender'])
df2
```

```python
# 3. 使用numpy创建df
# 3.1 创建ndarray对象
arr1 = np.arange(12).reshape(3, 4)
arr1
```

```python
# 3.2 将ndarray对象封装成dataframe类型
df3 = pd.DataFrame(data = arr1, index = ['a', 'b', 'c'], columns = ['一', '二', '三', '四'])
df3
```

```python
# 4. 生信息处理案例——dataframe入门案例
# 4.1 10名同学,5科成绩,范围40-100分
score_df = pd.DataFrame(data = np.random.randint(40, 101, (10, 5)))
score_df
```

```python
# 4.2 修改dataframe对象的 列名,索引值
column_names = ['语文', '数学', '英语', '政治', '体育']
# index_names = ['学生1', '学生2', '学生3', '学生4', '学生5']
index_names = ['同学' + str(i) for i in range(score_df.shape[0])]

# 4.3 具体修改动作
# score_df.columns = column_names
# score_df.index = index_names

# 需要用字典精确修改
# score_df.rename(columns = {0:'语文', 3:'AI'}, index = {0:'同学0', 1:'同学1', 5:'同学5'}, inplace = True)
score_df.rename(
    columns = {i:column_names[i] for i in range(score_df.shape[1])},
    index = {i:index_names[i] for i in range(score_df.shape[0])},
    inplace = True
)

# 打印
score_df
```
