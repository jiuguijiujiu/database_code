- 完成具体需求

```python
# 1. 找到租金最低，和租金最高的房子
# 思路1：排序
# house_df.sort_values(by = 'price').head(1)
# house_df.sort_values(by = 'price').tail(1)

# 思路2：筛选
# house_df[house_df.price == house_df.price.min()]
# house_df[house_df.price == house_df.price.max()]

# house_df.query('price == price.min() or price == price.max()')

# 思路3：
house_df.nlargest(1, 'price')       # 最大的n个，参数一：n个，参数二：排序字段。可理解为思路一简化版
house_df.nsmallest(1, 'price')      # 最小的n个，参数一：n个，参数二：排序字段。可理解为思路一简化版
```

```python
# 2. 找到最近新上的10套房源
house_df.sort_values(by = 'update_time', ascending=False).head(10)
```

```python
# 3. 查看所有更新时间
house_df.update_time.sort_values().unique()
```

```python
# 4. 查看看房人数的平均值, 最大值, 最小值
house_df.view_num.mean()
house_df.view_num.max()
house_df.view_num.min()
house_df.view_num.describe()
```

```python
# 5. 查看不同看房人数的房源数量，as_index = False 分组字段不作为行索引（默认为True)
# house_df.groupby('view_num', as_index=False).build_time.count()
house_df.groupby('view_num', as_index=False).agg({'build_time':'count'})    # 效果同上，这个好，通用

# 修改列名
tmp_df.columns = ['view_num', 'house_count']
tmp_df
```

```python
# 6. 查看房租价格的分布, 例如: 平均值, 标准差, 中位数...
house_df.price.mean()
house_df.price.std()
house_df.price.median()
```

```python
# 7. 找到看房人数最多的朝向
house_df.groupby('direction', as_index = False).agg({'view_num':'sum'}).query('view_num == view_num.max()')
```

```python
# 8. 查找最受欢迎的房型
house_df.groupby('house_type', as_index = False).agg({'view_num':'sum'}).query('view_num == view_num.max()')
```

```python
# 9. 查找房子的平均租房价格 （元/平米）
# 思路1：计算每套房子租房价格平均值，再求平均值之和的平均值
house_df['price/area'] = house_df.price / house_df.area
house_df['price/area'].mean()                 # 87.72268429900454

# 思路2：总租金/总面积
house_df.price.sum() / house_df.area.sum()    # 86.36679645198356
```

```python
# 10. 找到出租房源最多的小区
tmp_df = house_df.groupby('address', as_index = False).agg({'district':'count'}).query('district == district.max()')
tmp_df.columns = ['address', 'count']
tmp_df
```
