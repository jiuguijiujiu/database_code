```python
# 导包
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 1. 加载数据集, 并设置第一列为索引列
df = pd.read_csv('./数据集/tips.csv')          # 离线方式
# df = sns.load_dataset('tips')                 # 在线方式
df.head()
```

# 1. 直方图

```python
# 需求1：绘制图像，显示 男女 总账单的情况
# 1. 创建画布
fig, ax = plt.subplots(figsize = (10, 5))

# 2. 绘制图形
# 参数1：data 数据集
# 参数2：x x轴列名
# 参数3：hue 分组字段
sns.histplot(data = df, x = 'total_bill', hue = 'sex')

# 3. 添加标题
ax.set_title("男女总账单情况", fontsize = 15)

# 4. 展示图形
plt.show()
```

# 2. 散点图 scatter(), regplot(), joinplot()

```python
# 需求：描述 总账单 与 小费 的分布情况
# 1. 创建画布
fig, ax = plt.subplots(figsize = (10, 5))

# 2. 绘制图像
# 思路一：scatter()
# 参数1：data 数据集
# 参数2：x x轴列名
# 参数3：y 轴列名
# 参数4：hue 分组字段
# sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex')

# 思路二：regplot()，可以绘制散点图，还可以绘制拟合回归线（底层是线性回归）
# sns.regplot(data = df, x = 'total_bill', y = 'tip')     # 参数fig_reg = Ture（默认）是否显示拟合回归线
# sns.regplot(data = df, x = 'total_bill', y = 'tip', fit_reg = False)        # fit_reg = False，不显示拟合回归线

# 思路三：joinplot()，可以绘制散点图，还可以绘制拟合回归线（底层是线性回归）
sns.jointplot(data = df, x = 'total_bill', y = 'tip', kind = 'hex')     # kind = 'hex'  ——> 蜂巢图

# 3. 添加标题
ax.set_title("总账单 与 小费 的分布情况", fontsize = 15)

# 3. 展示图形
plt.show()
```

# 3. 箱线图 boxplot(), violinplot()

```python
# 绘制箱线图（也叫盒须图），展示 时间（time） 与 总金额（total_bill） 的分布情况
sns.boxplot(data = df, x = 'time', y = 'total_bill')
plt.show()
```

```python
# 需求：性别 与 总账单 ， 小费 的分布情况
# 小提琴图：violinplot() 适用于箱线图+密度图
# 参数1：data 数据集
# 参数2：x x轴列名
# 参数3：y 轴列名
# 参数4：hue 分组字段
# 参数5：是否箱线图分割，默认不分割
sns.violinplot(data = df, x = 'time', y = 'total_bill', hue = 'sex', split = True)
plt.show()
```
