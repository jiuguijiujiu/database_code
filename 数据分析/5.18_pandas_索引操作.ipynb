```python
import pandas as pd
```

```python
# 1. 加载数据
df = pd.read_csv('./数据集/stock_day.csv')
df
```

```python
# 2. 移除不需要的列
df.drop(columns = ['ma5', 'ma10', 'ma20', 'v_ma5', 'v_ma10', 'v_ma20'], inplace = True, axis = 0)
```

```python
# 3. 查看详细信息
print(df)
print(df.info())
print(df.describe())
```

# 索引操作

```python
# 1.  直接使用行列索引获取元素（先列后行）
df['open']['2018-02-23']
```

```python
# 2. 结合loc或者iloc使用索引获取数据
# 2.1 loc[]
df.loc['2018-02-23', 'open']
df.loc['2018-02-27':'2018-02-14', ['open', 'high']]
```

```python
# 2.2 iloc[]
df.iloc[0, 0]
df.iloc[0:5, 0:2]
```
