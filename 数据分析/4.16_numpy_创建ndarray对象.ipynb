# 1. 创建numpy的ndarray对象

## 1.1 array() ——> 把python列表转换为ndarray对象

```python
# 1. 创建列表对象
my_list = [1,2,3,4,5,6]
print(my_list)
print(type(my_list))
# 将 列表对象 转换为 ndarray对象
arr1 = np.array(my_list)
print(arr1)
print(type(arr1))
```

## 1.2 arange()函数创建ndarray对象

```python
# 1. 创建ndarray对象
# arange(起始, 结束, 步长, dtype = 类型) 类似于python中range()函数
arr2 = np.arange(0, 10, 2, dtype = np.float32)
print(arr2)
print(type(arr2))
print("元素类型：", arr2.dtype)
```

## 1.3 随机数生成ndarray对象

```python
# 1. random.rande()
arr3 = np.random.rand(3, 4)
print(arr3)
print(type(arr3))
```

```python
# 2. randow.randint()
arr4 = np.random.randint(1, 5, size = (3, 5))
print(arr4)
print(type(arr4))
```

```python
# 3. randow.uniform()
arr5 = np.random.uniform(1, 5, size = (3, 5))
print(arr5)
print(type(arr5))
```
