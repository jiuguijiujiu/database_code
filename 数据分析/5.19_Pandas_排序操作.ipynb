```python
import pandas as pd
```

```python
# 1. 加载数据
df = pd.read_csv('./数据集/stock_day.csv')
df
```

```python
# 2. 移除不需要的列
df.drop(columns = ['ma5', 'ma10', 'ma20', 'v_ma5', 'v_ma10', 'v_ma20'], inplace = True, axis = 0)
```

```python
# 3. 查看详细信息
print(df)
print(df.info())
print(df.describe())
```

# 1. 赋值操作

```python
# 1. 赋值操作
df['open'] = 23
df.low = 1
df
```

# 2. 排序操作

```python
# 2. 排序操作
# 2.1 按照1个列进行排序
df.sort_values(by = 'open', ascending = True)   # 升
df.sort_values(by = 'open', ascending = False)  # 降
```

```python
# 2.2 按照多个列进行排序
df.sort_values(by = ['open', 'high'], ascending = [False, False])
```

```python
# 2.2 对索引进行排序
df.sort_index(ascending = True)
```

# 3. series对象也有sort_index()，sort_values()排序

```python
# 3 series对象也有sort_index()，sort_values()排序
# df.open.sort_index(ascending = True)
df.open.sort_values(ascending = True)
```
