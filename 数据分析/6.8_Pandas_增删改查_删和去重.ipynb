```python
import pandas as pd
```

```python
# 1. 加载数据
df = pd.read_csv('./数据集/1960-2019全球GDP数据.csv', encoding='gbk')
df
```

```python
# 2. 拷贝一份原始数据出来进行操作
df2 = df[:10].copy()
df2
```

```python
# 3. 删除行——drop()
# df2.drop(index = [0, 2, 4])
df2.drop(index = 0, inplace = True)
```

```python
# 4. 删除列——del关键字，会直接修改源数据
del df2['year']
df2
```

```python
# 5. 删除列——drop()
df2.drop(columns=['country'], inplace = True)
df2
```

```python
# 6. 去重
# 场景1：dataframe去重——>以 行 做单位比较
df3 = df[:5].copy()

# 拼接 df5 + df5,形成又重复数据的dataframe
df4 = pd.concat([df3, df3])

# dataframe去重
df4.drop_duplicates(inplace = True)
df4
```

```python
# 场景2：series去重——>以 列 做单位比较
df4.country.drop_duplicates()
df.country.drop_duplicates()
```
