```python
import numpy as np
```

# 1. numpy去重函数

```python
# 1. 定义数组
arr1 = np.array([[1, 2, 1], [2, 3, 5]])
print(arr1)
```

```python
arr2 = np.unique(arr1)
print('arr1:', arr1)
print('arr2:', arr2)
```

# 2. numpy排序函数

```python
arr1 = np.array([[1, 2, 1], [2, 3, 5]])
print(arr1)
```

```python
arr2 = np.sort(arr1)
print('arr1:', arr1)
print('arr2:', arr2)

arr1.sort()
print('arr1:', arr1)
```
