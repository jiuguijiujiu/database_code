```python
import numpy as np
import pandas as pd
```

```python
# 4.1 10名同学,5科成绩,范围40-100分
score_df = pd.DataFrame(data = np.random.randint(40, 101, (10, 5)))
score_df
```

```python
# 4.2 修改dataframe对象的 列名,索引值
column_names = ['语文', '数学', '英语', '政治', '体育']
# index_names = ['学生1', '学生2', '学生3', '学生4', '学生5']
index_names = ['同学' + str(i) for i in range(score_df.shape[0])]

# 4.3 具体修改动作
# score_df.columns = column_names
# score_df.index = index_names

# 需要用字典精确修改
# score_df.rename(columns = {0:'语文', 3:'AI'}, index = {0:'同学0', 1:'同学1', 5:'同学5'}, inplace = True)
score_df.rename(
    columns = {i:column_names[i] for i in range(score_df.shape[1])},
    index = {i:index_names[i] for i in range(score_df.shape[0])},
    inplace = True
)

# 打印
score_df
```

```python
# 1. shape,维度,行列数
print(score_df.shape)

# 2. index,索引列
print(score_df.index)

# 3. columns,列名
print(score_df.columns)

# 04. values,数据
print(score_df.values)
```
