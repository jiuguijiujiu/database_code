```python
import numpy as np
import pandas as pd
```

```python
# 1. 创建dataframe对象
# 1. 字典+列表创建df，使用默认自增索引
info = {
    'name':['一一', '二二', '三三', '四四'],
    'gender':['男', '女', np.nan, '保密'],
    'age':[20, 30, 40, 50]
}

df1 = pd.DataFrame(data = info)
df1
```

```python
# 2. info
df1.info()
```

```python
# 3. datetime类型
df2 = pd.DataFrame(data = ['2021-1-1', '2021-1-2', '2021-1-3', '2021-1-4'], dtype = 'datetime64[ns]')
print(df2)
print(df2.dtypes)
```

```python
# 4. datedelta类型
start_date = pd.to_datetime('2002-6-7')
end_date = pd.to_datetime('2026-7-2')

print(end_date - start_date)
print(type(end_date - start_date))
```

```python
# 5. category类型
s1 = pd.Series(['男', '女', '保密'], dtype='category')
print(s1)
print(type(s1))
print(s1.dtype)
```
