```python
import pandas as pd
```

```python
# 1. 加载数据
df = pd.read_csv('./数据集/1960-2019全球GDP数据.csv', encoding='gbk')
df
```

```python
# 2. 获取前5条数据
df.head(5)
```

```python
# 3. 获取后3条数据
df.tail(3)
```

```python
# 4. 根据列名获取数据
df['year']      # series对象
df[['year']]    # dataframe对象
df.year         # series对象

df[['year', 'country']]    # dataframe对象
```

```python
# 5. 根据 行号 获取数据
df[:5]
df[0:5:2]
df[10::5]
```

```python
# 6. 通过query()函数，结合条件获取
# 需求：查询中美日三国2015——2019的数据
df.query('country in ["中国", "美国", "日本"] and (year >= 2015 and year <= 2019)')
```

```python
# 7. 排序
df.sort_index(ascending = False)
df.sort_values(by=['year', 'GDP'], ascending=[True, False])
```

```python
# 8. rank()函数 类似 sql窗口函数
df7 = pd.DataFrame({
'姓名': ['小明', '小美', '小强', '小兰'],
'成绩': [100, 90, 90, 80]
})

df7
```

```python
# rank()直接排名
df7.rank()

# 给df新增一列用于显示排名
df7['成绩排名_min'] = df7.成绩.rank(method='min', ascending=False)
df7['成绩排名_max'] = df7.成绩.rank(method='max', ascending=False)
df7['成绩排名_dense'] = df7.成绩.rank(method='dense', ascending=False)
df7['成绩排名_ave'] = df7.成绩.rank(method='average', ascending=False)
df7['成绩排名_默认'] = df7.成绩.rank(ascending=False)

df7
```
