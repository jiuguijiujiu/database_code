# 1. apply()函数，可执行自定义函数

```python
# 需求1：同时获取 多列 最大值与最小值的 差值
# df.open.max() - df.open.min()
# df.close.max() - df.close.min()

# 思路一：分解版
# 自定义函数my_func,来接收 某列数据，计算该列最大值与最小值，返回差值
def my_func(col):
    return col.max() - col.min()

# 获取open与close列
df[['open', 'close']]
df.loc[:, ['open', 'close']]
df.iloc[:, [0, 2]]
df.iloc[:, 0:3:2]

# 使用apply()函数
df[['open', 'close']].apply(my_func, axis = 0)
```

```python
# 思路二： 合并版，通过lambda函数实现
df[['open', 'close']].apply(lambda col: col.max() - col.min())
```
