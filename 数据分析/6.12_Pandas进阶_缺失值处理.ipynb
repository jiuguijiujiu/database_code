```python
import numpy as np
import pandas as pd
```

# 1. 删除缺失值

```python
# 1. 读取数据
movie_df = pd.read_csv('./数据集/movie.csv')
movie_df
```

```python
# 2. 查看下数据的介绍
movie_df.columns
movie_df.info()
movie_df.describe()
```

```python
# 3.删除缺失值
movie_df.dropna(inplace=False, axis=0)
```

# 2. 填充缺失值

```python
# 1. 查看数据
movie_df
```

```python
# 2. 判断 某列的 某个值 是否为缺失值
pd.isnull(movie_df)
pd.notnull(movie_df)

# 3. 判断某列是否是 包含缺失值的列
np.all(pd.notnull(movie_df))
```

```python
# 4. 填充缺失值
# 写法1：填充固定值
movie_df.fillna(23).info()
```

```python
# 写法2：填充每列平均值
movie_df['Revenue (Millions)'].fillna(movie_df['Revenue (Millions)'].mean(), inplace=True)
movie_df['Metascore'].fillna(movie_df['Metascore'].mean(), inplace=True)
movie_df.info()
```

```python
# 5. for循环的方式，使用 每列平均值 来填充 每列的缺失值
# 5.1 获取每列 列名
for col_name in movie_df.columns:
    # 5.2 判断 某列 是否有缺失值
    if not np.all(pd.notnull(movie_df[col_name])):
        # 5.3 说明该列有缺失值，填充缺失值
        movie_df[col_name].fillna(movie_df[col_name].mean(), inplace=True)
```
