```python
import pandas as pd

import matplotlib as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

# 1. 绘制中美日三国gdp趋势图--带图例--英文

```python
# 1. 获取数据
df = pd.read_csv('./数据集/1960-2019全球GDP数据.csv', encoding='gbk')
```

```python
# 2. 加载三国数据
chain_df = df[df.country == '中国'].set_index('year')
usa_df = df[df.country == '美国'].set_index('year')
jp_df = df[df.country == '日本'].set_index('year')
chain_df
```

```python
# 3. 分别修改中美日的 gdp字段 改成 对应国家字段
chain_df.rename(columns={'GDP':'chain'}, inplace=True)
usa_df.rename(columns={'GDP':'usa'}, inplace=True)
jp_df.rename(columns={'GDP':'jp'}, inplace=True)
chain_df
```

```python
# 4. 绘制折线图(分开了)
chain_df.plot()
usa_df.plot()
jp_df.plot()
```

```python
# 5. 三表合一,且有图例--英文
chain_df.chain.plot(legend=True)            # legend=True,设置图例, 不写没有图例
usa_df.usa.plot(legend=True)
jp_df.jp.plot(legend=True)
```
