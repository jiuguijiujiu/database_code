```python
# 导包
import numpy as np
```

# 1. numpy属性介绍

## 1.1 创建ndarray对象

```python
# 1. 创建numpy的ndarray对象
arr1 = np.arange(15).reshape(3, 5)
print(arr1)
```

## 1.2演示numpy常用属性

```python
print(f"numpy的轴：{arr1.ndim}")
print(f"numpy的维度：{arr1.shape}")
print(f"numpy的元素类型：{arr1.dtype}")
print(f"numpy的元素个数：{arr1.size}")
print(f"numpy的元素占用字节数：{arr1.itemsize}")
print(f"numpy的类型：{type(arr1)}")
```
