# 2. 逻辑运算符 &，|

```python
# 2.1 查看源数据
df
```

```python
# 2.2筛选 open > 23
df[df.open > 23]

# 2.3 筛选 open > 23 且  < 24
df[(df.open > 23) & (df.open < 24)]
```

```python
# query()函数优化：筛选 open > 23 且  < 24
df.query('open > 23 & open < 24')
```

# 3. 固定值的筛选isin()

```python
# 3.1 查询 open 为 23.53 与 23.67 的数据
df[(df.open == 23.53) | (df.open == 23.67)]
df[df.open.isin([23.53, 23.67])]


df.query('open == 23.53 | open == 23.67')
df.query('open in [23.53, 23.67]')
```
