```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# 1. 准备数据
data = pd.read_csv('./数据集/csv示例文件.csv', encoding='gbk', index_col = 0)
data
```

```python
# 2. 导包
# from sqlalchemy import create_engine
```

```python
# 3. 创建引擎对象
engine = create_engine('mysql+pymysql://root:123456@localhost:3306/day01?charset=utf8')

# 4. 写入数据
data.to_sql('my_table', engine, index = False, if_exists = 'replace')
print("写入成功")
```

```python
# 5. 读取sql表中数据
sql_df = pd.read_sql('my_table', engine)    # 读全表
sql_df = pd.read_sql('select name, AKA from my_table limit 0,2', engine)    # 读指定
sql_df
```
