```python
import pandas as pd
```

```python
# 1. 准备数据
df = pd.read_csv('./数据集/1960-2019全球GDP数据.csv', encoding='gbk')
df

df1 = df[:10]
df1

df2 = df[10:20]
df2
```

```python
# 2. concat()合并数据
new_df = pd.concat([df1, df2], axis=0)  # aix = 0，列合并，默认
new_df = pd.concat([df1, df2], axis=1)  # aix = 1，行合并
new_df
```

```python
# 3. 修改df2行索引
df2.index = [0, 1, 2, 3 ,4 , 10, 11, 12, 13, 14]
df2
```

```python
# 4. df1 与 df2 再次行合并
new_df = pd.concat([df1, df2], axis=1)  # aix = 1，行合并
new_df
```

```python
# 5. df1 与 df2 再次行合并
# aix = 1，行合并，默认outer满外连接
new_df = pd.concat([df1, df2], axis=1, join = 'outer')

# aix = 1，行合并，指定inner内连接
new_df = pd.concat([df1, df2], axis=1, join = 'inner')

# aix = 1，行合并，指定left左外连接，但是不支持
new_df = pd.concat([df1, df2], axis=1, join = 'left')
new_df
```
