```python
import pandas as pd
```

```python
# 1. 加载数据
df = pd.read_csv('./数据集/1960-2019全球GDP数据.csv', encoding='gbk')
df
```

```python
# 2. 拷贝一份原始数据出来进行操作
df2 = df[:5].copy()
df2
```

```python
# 3. 新增列
# 思路一：通过直接赋值的方式直接添加新列
df2['c1'] = 23
df2['c2'] = [1, 2, 3, 4, 5]
df2['c3'] = df2.year * 2

def my_func1():
    return 250

df2['c4'] = my_func1()

df2
```

```python
# 4. 思路2：通过assign（）的方式新增列
df3 = df[:5].copy()
df3
```

```python
# 增加1列
df3.assign(c1 = 23)
# 增加多列
df3.assign(
    c1 = 10,
    c2 = [1, 2, 3, 4, 5],
    c3 = df3.year * 2,
    c4 = my_func1()

)
```
