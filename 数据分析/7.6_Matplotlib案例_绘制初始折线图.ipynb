```python
import matplotlib.pyplot as plt
import random

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 1. 准备 x轴（时间） 与 y轴（温度） 数据
x = range(61)
y = [random.uniform(15, 18) for i in x]

# 2. 创建画布
plt.figure(figsize=(10,5), dpi=100)

# 3. 绘图
plt.plot(x, y, color='blue')

# 4. 展示绘图结果
plt.show()
```

# 1. 添加x,y刻度信息和网格

```python
# 1. 准备 x轴（时间） 与 y轴（温度） 数据
x = range(60)
y = [random.uniform(15, 18) for i in x]

# 2. 创建画布
plt.figure(figsize=(10,5), dpi=100)

# 3. 绘图
plt.plot(x, y, color='red')

# 4. 添加x,y刻度信息
# 4.1 准备x轴与y轴刻度信息
x_ticks_label = [f'11点{i}分' for i in x]
y_ticks = range(40)

# 4.2 添加x轴，y轴
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 5. 添加网格
plt.grid(linestyle='-.', alpha=0.5)

# 展示绘图结果
plt.show()
```

# 2. 添加标题

```python
# 1. 准备 x轴（时间） 与 y轴（温度） 数据
x = range(60)
y = [random.uniform(15, 18) for i in x]

# 2. 创建画布
plt.figure(figsize=(10,5), dpi=100)

# 3. 绘图
plt.plot(x, y, color='red')

# 4. 添加x,y刻度信息
# 4.1 准备x轴与y轴刻度信息
x_ticks_label = [f'11点{i}分' for i in x]
y_ticks = range(40)

# 4.2 添加x轴，y轴
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 5. 添加网格
plt.grid(linestyle='-.', alpha=0.5)

# 6. 添加标题
plt.title('北京11点——12点温度', fontsize=20)      # fontsize控制字体大小

# 7. 添加x，y轴标签
plt.xlabel('时间', fontsize=10)
plt.ylabel('温度', fontsize=10)

# 8. 保存图片
plt.savefig('./数据集/my_picture.png')

# 展示绘图结果
plt.show()
```
