```python
import numpy as np
import pandas as pd
```

```python
# 1. 创建series对象,默认自增索引
s1 = pd.Series([1,2,3,4,5,6])
print(s1)
```

```python
# 2. 创建series对象,自定义索引
s2 = pd.Series([1,2,3,4,5,6], index=['a','b','c','d','e','f'])
print(s2)
```

```python
# 3. 使用字典或者元组创建series对象
# 元组
s3 = pd.Series((1,2,3,4,5,6))
print(s3)

# 字典
s4 = pd.Series({'a':1,'b':2,'c':3, 'd':4, 'e':5, 'f':6})
print(s4)
```

```python
# 4. 使用numpy创建series对象
s5 = pd.Series(np.array([1,2,3,4,5,6]))
print(s5)
```
