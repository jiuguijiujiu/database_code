```python
import numpy as np
```

# 1. numpy基本运算

```python
arr1 = np.array([11, 22, 33])
arr2 = np.array([1, 2, 3])
print(f'arr1:{arr1}')
print(f'arr2:{arr2}')
# 对应元素做运算
print(f'arr1 + arr2: {arr1 + arr2}')
print(f'arr1 - arr2: {arr1 - arr2}')
print(f'arr1 * arr2: {arr1 * arr2}')
print(f'arr1 / arr2: {arr1 / arr2}')
```

# 2. numpy矩阵乘法

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(f'arr1:{arr1}')
print(f'arr2:{arr2}')
```

```python
# 对应元素相乘(必须行列数一致)
print(f'arr1 * arr2: {arr1 * arr2}')
print(f'multiply(arr1, arr2): {np.multiply(arr1, arr2)}')
```

```python
# 点积运算(按照矩阵乘法规则)
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[1, 2], [3, 4], [5, 6]])
print(f'arr1:{arr1}')
print(f'arr2:{arr2}')
print(f'arr1 @ arr2: {arr1.dot(arr2)}')
print(f'arr1 @ arr2: {np.dot(arr1, arr2)}')
print(f'arr1 @ arr2: {arr1 @ arr2}')            # 语法糖
```
